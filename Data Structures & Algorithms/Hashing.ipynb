## Names

```python
# python3

class Query:

    def __init__(self, query):
        self.type = query[0]
        if self.type == 'check':
            self.ind = int(query[1])
        else:
            self.s = query[1]


class QueryProcessor:
    _multiplier = 263
    _prime = 1000000007

    def __init__(self, bucket_count):
        self.bucket_count = bucket_count
        # store all strings in one list
        self.elems = [[] for i in range(bucket_count)]

    def _hash_func(self, s):
        ans = 0
        for c in reversed(s):
            ans = (ans * self._multiplier + ord(c)) % self._prime
        return ans % self.bucket_count

    def write_search_result(self, was_found):
        print('yes' if was_found else 'no')

    def write_chain(self, chain):
        print(' '.join(chain))

    def read_query(self,i):
        queries = ['check 0',
                    'find help',
                    'add help',
                    'add del',
                    'add add',
                    'find add',
                    'find del',
                    'del del',
                    'find del',
                    'check 0',
                    'check 1',
                    'check 2',
                  ]
        return Query(queries[i].split())
    
    def process_queries(self):
        #n = int(input())
        for i in range(12):
            self.process_query(self.read_query(i))
    def process_query(self, query):
        #print(self.elems)
        if query.type == "check":
            # use reverse order, because we append strings to the end
            chain = self.elems[query.ind]
            self.write_chain(cur for cur in reversed(chain))
        else:
            hash_num = self._hash_func(query.s)
            chain = self.elems[hash_num]

            try:
                ind = chain.index(query.s)
            except ValueError:
                ind = -1

            if query.type == 'find':
                self.write_search_result(ind != -1)
            elif query.type == 'add':
                if ind == -1:
                    chain.append(query.s)
            else:
                if ind != -1:
                    chain.pop(ind)



if __name__ == '__main__':
    bucket_count = 3
    proc = QueryProcessor(bucket_count)
    proc.process_queries()
```
